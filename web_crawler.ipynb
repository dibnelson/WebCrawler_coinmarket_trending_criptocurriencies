```python
import requests
from requests.exceptions import HTTPError
from bs4  import BeautifulSoup
import re
import csv
import pandas as pd
```

```python
URL = 'https://coinmarketcap.com/pt-br/trending-cryptocurrencies/'
content = None
def webXtractor(url : str):
    try:
        out = requests.get(url)
        out.raise_for_status()
    except HTTPError as exc:
        print(exc)
    else:
        content = out.text
        return content
    
```

```python
content = webXtractor(URL)
```

```python
with open(file='cripto.html',mode='w',encoding='utf8')as fl:
    fl.write(content)
```

```python
page = BeautifulSoup(open('cripto.html',mode='r',encoding='utf8'),'html.parser')
table = page.find('tbody')
extractor = []

for line in table.find_all('tr'):
    text_column = list()
    for column in line.find_all('td'):
        text = column.get_text().strip().split('\n')
        text_column += text
    extractor.append(text_column)
        
```

```python
with open(file='./cripto.csv',mode='w',encoding='utf8') as fl:
    write = csv.writer(fl,delimiter=';')
    write.writerow(['#','name','price','24h','7d','30d','market'])
    for line in extractor:
        write.writerow([str(line[1]),str(line[2]),str(line[3]),str(line[4]),str(line[5]),str(line[6]),str(line[7])])
```

```python
cripto_df = pd.read_csv('./cripto.csv',sep=';')
```

```python
def cleaner (DataFrame,column):
    for col in column:
        DataFrame[col] = DataFrame[col].str.replace(r'\d+',' - ',regex=True)
```

```python
cleaner(cripto_df,['name'])
```
